## Distribuição de Médias

```r
library(ggplot2)
library(dplyr)
simulacoes <- 10000
amostra <- c(10,20,30)
set.seed(123)
```

```r
#Criando Matriz Exemplo
exemplo <- matrix(sample(0 : 1, 10 * amostra[1], replace = TRUE),
                           10)
exemplo
```

```r
#Calculando a média nas linhas
apply(exemplo,1,mean)
```

```r
dat <- data.frame(
        #Lançamentos simples
        x = c(sample(0 : 1, simulacoes, replace = TRUE),
        #Médias de lançamentos para primeiro elemento de amostra
        apply(matrix(sample(0 : 1, simulacoes * amostra[1], replace = TRUE),
        simulacoes), 1, mean),
        #Médias de lançamentos para segundo elemento de amostra
        apply(matrix(sample(0 : 1, simulacoes * amostra[2], replace = TRUE),
        simulacoes), 1, mean),
        #Médias de lançamentos para terceiro elemento de amostra
        apply(matrix(sample(0 : 1, simulacoes * amostra[3], replace = TRUE),
        simulacoes), 1, mean)
        ),
        #fatores
        tamanho = factor(rep(c(1, amostra[1], amostra[2], amostra[3]),
                          rep(simulacoes, 4)))
        )
 mu <- dat %>% group_by(tamanho) %>% summarise(media=mean(x),.groups="drop")
mu
```

```r
g <- ggplot(dat, aes(x = x, fill = tamanho)) +
        geom_histogram( binwidth = 1/12)+
        geom_density(alpha=0.2)+ facet_grid(. ~ tamanho)+
        geom_vline(data=mu, aes(xintercept=media),
                   linetype="dashed")+
        labs( x="Média das amostras",y = "Número de lançamentos",fill="Tamanho \n amostra")

g
```

```r
dat <- data.frame(
        #Lançamentos simples
        x = c(sample(1:6, simulacoes, replace = TRUE),
        #Médias de lançamentos para primeiro elemento de amostra
        apply(matrix(sample(1:6, simulacoes * amostra[1], replace = TRUE),
        simulacoes), 1, mean),
        #Médias de lançamentos para segundo elemento de amostra
        apply(matrix(sample(1:6, simulacoes * amostra[2], replace = TRUE),
        simulacoes), 1, mean),
        #Médias de lançamentos para terceiro elemento de amostra
        apply(matrix(sample(1:6, simulacoes * amostra[3], replace = TRUE),
        simulacoes), 1, mean)
        ),
        #fatores
        tamanho = factor(rep(c(1, amostra[1], amostra[2], amostra[3]),
                          rep(simulacoes, 4)))
        )
 mu <- dat %>% group_by(tamanho) %>% summarise(media=mean(x),.groups="drop")
mu
```

```r
g <- ggplot(dat, aes(x = x, fill = tamanho)) +
        geom_histogram( binwidth = 1/12, colour = "black")+
        geom_density(alpha=0.2)+ facet_grid(. ~ tamanho)+
        geom_vline(data=mu, aes(xintercept=media),
                   linetype="dashed")+
        labs( x="Média das amostras",y = "Número de lançamentos",fill="Tamanho \n amostra")

g
```
